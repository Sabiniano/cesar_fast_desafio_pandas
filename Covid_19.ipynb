### 3. Crie uma nova coluna chamada ‘’tipo_faixa_etaria”, onde deverá conter os seguintes valores:
- Crianças: 3 a 12 anos
- Adolescentes: 13 a 19 anos
- Jovens adultos: 20 a 39 anos
- Adultos de meia-idade: 40 a 59 anos
- Idosos: 60 anos ou mais

### R. Inicialmente é necessário fazer um tratamento na base. Para criação desta nova coluna, iremos trabalhar com o campo "idade", então, devemos verificar se tem linhas idades vazias. Encontramos 6 registros sem a informação de idade, optamos por apaga-la, pois por ser um número muito pequeno, não influi no resultado.  Após isso, constatamos que o campo idade era um "object", portanto, transformamos ele para númerico. Após conclusão dos procedimentos acima, criamos uma função para criar a nova coluna "tipo_faixa_etaria", conforme solicitado nessa questão.

O método isna() mostra os valores nulos com True e os não nulos como False

```python
df_empty = df.isna()
df_empty
```

Acrescentando o método any() mostra as colunas que tem pelo menos um campo nulo

```python
df_empty = df.isna().any()
df_empty
```

Acrescentando o método sum(), pode-se ver quantos campos nulos tem em cada coluna

```python
df_empty = df.isna().sum()
print(df_empty)
```

O método dropna() elimina todos os campos nulos. Neste caso, eliminamos todos os 6 valores vazios da coluna "idade".  Como 6 é um valor muito pequeno para o tamanho da base, não influi no resultado.

```python
df = df.dropna(subset=['idade'])
df_subset = df.isna().any()
df_subset
```

```python
df_subset = df.isna().sum()
print(df_subset)
```

Com a propriedade dtype, verifica-se que a coluna "idade" é object e é necessário transforma-la em numerica

```python
df['idade'].dtype
```

Transformando a coluna "idade" para numerica

```python
df['idade'] = pd.to_numeric(df['idade'], errors='coerce', downcast='integer')
df['idade'].head(2)
```

Verifica-se agora que a coluna "idade" é numerica

```python
df['idade'].dtype
```

```python
def TipoFaixaEtaria(idade):
    if idade >= 60:
        return "Idosos"
    elif idade >= 40:
        return "Adultos de meia idade"
    elif idade >= 20:
        return "Jovens Adultos"
    elif idade >= 13:
        return "Adolescentes"
    elif idade >= 3:
        return "Crianças"
    else:
        return "Indefinido"

df["tipo_faixa_etaria"] = df['idade'].apply(lambda x:  TipoFaixaEtaria(x))
df.head(5)
```

### 4. Crie uma nova coluna chamada ‘’regiao”, onde deverá conter valores:
- Metropolitana: Recife, Olinda, Jaboatão dos Guararapes, Paulista, Camaragibe, São Lourenço da Mata, Abreu e Lima, Igarassu, Cabo de Santo Agostinho
- Interior: Cidades que não estão inclusas em Metropolitana


```python
def Regiao(municipio):
    metropolitanas = ['RECIFE', 'JABOATAO DOS GUARARAPES','OLINDA', 'PAULISTA', 'CAMARAGIBE', 'SAO LOURENCO DA MATA', 'ABREU E LIMA', 'IGARASSU', 'CABO DE SANTO AGOSTINHO']

    if municipio in metropolitanas:
        return 'METROPOLITANA'
    else:
        return 'INTERIOR'

df['regiao'] = df['municipio'].apply(lambda x: Regiao(x))
df.head(5)
```

### 5. Conte quantas pessoas receberam a vacina para cada cada  ’regiao” e cada ‘’tipo_faixa_etaria”

```python
df[['regiao', 'tipo_faixa_etaria']].groupby('regiao')['tipo_faixa_etaria'].value_counts()
```

### 6. Determine a idade média das pessoas vacinadas para  cada  ’regiao” e cada ‘’tipo_faixa_etaria”

```python
df.groupby(['regiao', 'tipo_faixa_etaria'])['idade'].mean().reset_index(name='idade_media')
```

### 7. Identifique e exiba o tipo de vacina mais administrado para cada  ’regiao” e cada ‘’tipo_faixa_etaria”

```python
df.groupby(['regiao'])[['tipo_faixa_etaria', 'vacina_fabricante' ]].value_counts()
```
